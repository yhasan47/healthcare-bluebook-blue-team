```python
provider_chunks = pd.read_csv(
    PROVIDER,
    sep='\t',
    chunksize= 100000,
    usecols=['npi','nppes_provider_last_org_name',
             'nppes_provider_first_name','nppes_provider_mi','nppes_credentials' ,
             'nppes_entity_code','place_of_service','provider_type',
             'nppes_provider_zip','nppes_provider_city','nppes_provider_state',
             'hcpcs_code','hcpcs_description','bene_unique_cnt',
             'average_Medicare_allowed_amt','average_Medicare_payment_amt'],
    dtype = {'npi':'str','nppes_provider_zip': 'str'}
)
```

```python
#def filter_data(column):
filtered_dataset = []
for i, chunk in enumerate(provider_chunks):
    print(i +1)
    filtered_dataset.append(chunk[chunk['provider_type'] == 'Ambulatory Surgical Center'])
```

```python
dataset =pd.concat(filtered_dataset)
type(dataset)
```

```python
dataset.info()
```

```python
dataset_reset = dataset.reset_index(drop=True)
```
